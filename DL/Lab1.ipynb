Problem 1.2 (Applied Business Problem)

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

X = np.array([
    [0.8, 0.7],
    [0.2, 0.1],
    [0.7, 0.6],
    [0.1, 0.3],
    [0.6, 0.8],
    [0.3, 0.2],
    [0.9, 0.5],
    [0.2, 0.4]
])

y = np.array([1,0,1,0,1,0,1,0])

model = LogisticRegression(C=1.0, solver='lbfgs', random_state=42)
model.fit(X, y)

print("Coefficients:", model.coef_[0])
print("Intercept:", model.intercept_[0])
print("Odds ratios:", np.exp(model.coef_[0]))
```

b) Predicted probabilities (threshold = 0.5)

```python
import pandas as pd

probs = model.predict_proba(X)[:,1]
preds = (probs >= 0.5).astype(int)

table = pd.DataFrame({
    "Customer": range(1, 9),
    "Actual": y,
    "P(churn)": probs.round(4),
    "Predicted": preds
})

table
```

c) Metrics

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

print(confusion_matrix(y, preds))
print(classification_report(y, preds))
print("ROC-AUC:", roc_auc_score(y, probs))
```

Problem 1.3 (Business Decision Analysis)

```python
cost_offer = 50
lost_revenue = 500
success_rate = 0.4
```

Expected Value

```python
p_churn = 0.7

expected_value = p_churn * (success_rate * lost_revenue - cost_offer)
expected_value
```

Expected value = $105 -> offering retention is profitable

b) Optimal probability threshold

```python
optimal_threshold = cost_offer / (success_rate * lost_revenue)
optimal_threshold
```

c) Reclassification at a new threshold

```python
import numpy as np

new_preds = (probs >= optimal_threshold).astype(int)

new_preds
```

```python
import pandas as pd

comparison = pd.DataFrame({
    "Customer": range(1, 9),
    "P(churn)": probs.round(4),
    "Predicted (0.5)": (probs >= 0.5).astype(int),
    "Predicted (0.25)": new_preds
})

comparison
```

Using cost-based analysis, the optimal probability threshold was found to be 0.25.
Compared to the default threshold of 0.5, this approach identifies more at-risk customers and leads to higher expected profit.
